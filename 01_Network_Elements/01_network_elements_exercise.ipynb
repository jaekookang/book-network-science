```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1, 2)
```

```python
G.nodes()
```

```python
G.edges()
```

```python
for n in G.neighbors(2):
    print(n)
```

```python
for n in G.nodes:
    print(n, G.neighbors(n))
```

```python
D = nx.DiGraph()
D.add_edge(1, 2)
D.add_edge(2, 1)
```

```python
D.number_of_nodes()
```

```python
D.number_of_edges()
```

```python
D.neighbors(2)
```

```python
for n in D.predecessors(2):
    print(n)
```

```python
for n in D.successors(2):
    print(n)
```

```python
B = nx.complete_bipartite_graph(4, 5)
C = nx.cycle_graph(4)
P = nx.path_graph(5)
S = nx.star_graph(6)
```

```python
fig, arr = plt.subplots(4, 1, facecolor='white', figsize=(6,10))
for ax, graph in zip(arr, [B, C, P, S]):
    nx.draw(graph, ax=ax)
    ax.set_axis_on()
    ax.set_xlabel(f'nodes={graph.degree()}')
```

## Weighted networks

```python
W = nx.Graph()
W.add_edge(1, 2, weight=6)
W.add_weighted_edges_from([(2,3,3), (2,4,5)])
```

```python
nx.draw(W, with_labels=True)
```

```python
W.edges(data='weight')
```

```python
matrix = nx.adjacency_matrix(W)
```

```python
matrix.data
```

```python
for n, neighbors in W.adjacency():
    print(n, neighbors)
```

```python
# Write edges
nx.write_edgelist(W, 'file.edges')
W2 = nx.read_edgelist('file.edges')
```

```python
nx.write_weighted_edgelist(W, 'wf.edges')
```
